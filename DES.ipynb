```python
def PC1(key):
    table = [[57,49,41,33,25,17,9],
            [1,58,50,42,34,26,18],
            [10,2,59,51,43,35,27],
            [19,11,3,60,52,44,36],
            [63,55,47,39,31,23,15],
            [7,62,54,46,38,30,22],
            [14,6,61,53,45,37,29],
            [21,13,5,28,20,12,4]]
    
    PC1_key = ""
    R,C = len(table) , len(table[0]) # number of rows , number of columns
    for i in range(R):
        for j in range(C):
            PC1_key += key[table[i][j] - 1]
            
    return PC1_key
```

```python
def LCS(key , round_number):
    table = [1,1,2,2,2,2,2,2,1,2,2,2,2,2,2,1]
    LCS_key = key[table[round_number]:len(key)] + key[0:table[round_number]]

    return LCS_key
```

```python
def PC2(key):
    table = [[14,17,11,24,1,5],
            [3,28,15,6,21,10],
            [23,19,12,4,26,8],
            [16,7,27,20,13,2],
            [41,52,31,37,47,55],
            [30,40,51,45,33,48],
            [44,49,39,56,34,53],
            [46,42,50,36,29,32]]
    
    PC2_key = ""
    R,C = len(table) , len(table[0])
    for i in range(R):
        for j in range(C):
            PC2_key += key[table[i][j] - 1]
            
    return PC2_key
```

```python
def keyGenerator(left_part , right_part , round_number):
    LCS_Lk = LCS(left_part , round_number)
    LCS_Rk = LCS(right_part , round_number)
    LCS_key = LCS_Lk + LCS_Rk
    PC2_key = PC2(LCS_key)
    
    return PC2_key , LCS_Lk , LCS_Rk
```

```python
def DES_KEY(k):
    key = k
    bin_key = (bin(int(key , 16))[2:]).zfill(len(key)*4) #64-bit key

    #Permutation Choice-1
    PC1_key = PC1(bin_key) #56-bit key

    Lk = PC1_key[0 : int(len(PC1_key)/2)] 
    Rk = PC1_key[int(len(PC1_key)/2) : len(PC1_key)] 

    key = []
    PC2_key , LCS_Lk , LCS_Rk = keyGenerator(Lk , Rk , 0)
    key.append(PC2_key)

    for i in range(1 , rounds_number):
        Lk = LCS_Lk
        Rk = LCS_Rk
        PC2_key , LCS_Lk , LCS_Rk = keyGenerator(Lk , Rk , i)
        key.append(PC2_key)
        
    return key
```

```python
def Inv_IP(plain_text):
    table = [[40, 8, 48, 16, 56, 24, 64, 32],
        [39, 7, 47, 15, 55, 23, 63, 31],
        [38, 6, 46, 14, 54, 22, 62, 30],
        [37, 5, 45, 13, 53, 21, 61, 29],
        [36, 4, 44, 12, 52, 20, 60, 28],
        [35, 3, 43, 11, 51, 19, 59, 27],
        [34, 2, 42, 10, 50, 18, 58, 26],
        [33, 1, 41, 9, 49, 17, 57, 25]]
    
    inv_pt = ""
    R,C = len(table) , len(table[0]) # number of rows , number of columns
    for i in range(R):
        for j in range(C):
            inv_pt += plain_text[table[i][j] - 1]
            
    return inv_pt
```

```python
def swap(plain_text):
    Lp = plain_text[0:32]
    Rp = plain_text[32:64]
    
    return Rp+Lp
```

```python
def IP(plain_text):
    table = [[58,50,42,34,26,18,10,2],
            [60,52,44,36,28,20,12,4],
            [62,54,46,38,30,22,14,6],
            [64,56,48,40,32,24,16,8],
            [57,49,41,33,25,17,9,1],
            [59,51,43,35,27,19,11,3],
            [61,53,45,37,29,21,13,5],
            [63,55,47,39,31,23,15,7]]
    
    IP_pt = ""
    R,C = len(table) , len(table[0]) # number of rows , number of columns
    for i in range(R):
        for j in range(C):
            IP_pt += plain_text[table[i][j] - 1]
            
    return IP_pt
```

```python
def round_function(plain_text , key):
    Lp = plain_text[0 : int(len(plain_text)/2)] #32-bit left part
    Rp = plain_text[int(len(plain_text)/2) : len(plain_text)] #32-bit right part

    #Expansion Permutation
    EP_pt = EP(Rp) #48-bit plain text
    
    #XOR the expanded plain text with the key
    XORD_pt = xor(EP_pt , key) #48-bit plain text
    
    #S-Box the xor'd plain text
    SBOX_pt = SBOX(XORD_pt) #32-bit plain text
    
    #Permute the S-Box output plain text
    permuted_pt = P(SBOX_pt) #32-bit plain text
    
    #XOR the permuted plain text with the left part of the plain text
    XORD_pt_Lpt = xor(permuted_pt , Lp)
    
    Rp_n = XORD_pt_Lpt
    Lp_n = Rp
    round_n_pt = Lp_n + Rp_n
    
    return round_n_pt
```

```python
def EP(plain_text):
    table = [[32,1,2,3,4,5],
            [4,5,6,7,8,9],
            [8,9,10,11,12,13],
            [12,13,14,15,16,17],
            [16,17,18,19,20,21],
            [20,21,22,23,24,25],
            [24,25,26,27,28,29],
            [28,29,30,31,32,1]]
    
    EP_pt = ""
    R,C = len(table) , len(table[0]) # number of rows , number of columns
    for i in range(R):
        for j in range(C):
            EP_pt += plain_text[table[i][j] - 1]
            
    return EP_pt
```

```python
def xor(plain_text , key):
    _xormap = {('0', '1'): '1', ('1', '0'): '1', ('1', '1'): '0', ('0', '0'): '0'}
    res = ''.join([_xormap[x, y] for x, y in zip(plain_text, key)])
    
    return res
```

```python
def P(plain_text):
    table = [[16,7,20,21],
            [29,12,28,17],
            [1,15,23,26],
            [5,18,31,10],
            [2,8,24,14],
            [32,27,3,9],
            [19,13,30,6],
            [22,11,4,25]]
    
    permuted_pt = ""
    R,C = len(table) , len(table[0]) # number of rows , number of columns
    for i in range(R):
        for j in range(C):
            permuted_pt += plain_text[table[i][j] - 1]
            
    return permuted_pt
```

```python
def DES_MSG(plain_text , key):
    pt = plain_text
    bin_pt = (bin(int(pt , 16))[2:]).zfill(len(pt)*4) #64-bit plain text
    IP_pt = IP(bin_pt)
    round_pt = []
    round_pt.append(round_function(IP_pt , key[0]))

    for i in range(1 , rounds_number):
        round_pt.append(round_function(round_pt[i - 1] , key[i]))

    swapped_pt = swap(round_pt[15])
    bin_cipher_text =  Inv_IP(swapped_pt)
    hex_cipher_text = hex(int(bin_cipher_text,2))
    
    return hex_cipher_text[2:len(hex_cipher_text)]
```
