```python
import numpy as np
```

```python
def readTextFile(path):
    plaintext_file = open(path,"r")
    plaintext = plaintext_file.read().split("\n")
    plaintext_file.close()
    return plaintext
```

```python
def unique(sequence):
    seen = set()
    return [x for x in sequence if not (x in seen or seen.add(x))]
```

```python
def keyMatrix(key , n):
    check = 0
    mat = np.zeros((n,n) , dtype=str)
    letter = unique(key)
    pos = 0
    m = 97
    i = 0
    while i < n:
        j = 0
        while j < n:
            if(pos < len(letter)):
                mat[i][j] = list(letter)[pos]
                pos += 1
                j += 1
            else:
                if( 
                ((check == 0) and (m == 105) and (any('i' in sublist for sublist in mat)) and not(any('j' in sublist for sublist in mat)))
                   or
                ((check == 0) and (m == 105) and not(any('i' in sublist for sublist in mat)) and (any('j' in sublist for sublist in mat)) )
                   or
                ((check == 0) and (m == 105) and (any('i' in sublist for sublist in mat)) and (any('j' in sublist for sublist in mat)) )
                  ):
                    m = 107
                    check = 1
                    
                elif( (check == 0) and (m == 105) and not(any('i' in sublist for sublist in mat)) and not(any('j' in sublist for sublist in mat)) ):
                    m = 105
                    check = 1
                    
                     
                alpha_letter = chr(m)
                if(not(any(alpha_letter in sublist for sublist in mat))):
                    mat[i][j] = alpha_letter
                    j += 1
                else:
                    j = j
                if(m == 105):    
                    m += 2
                else:
                    m += 1
            
        i += 1
            
                    
    return mat
```

```python
def playFairCipher(part , keyMatrix):
    cpart = list("  ")

    if((part[0] == 'i') and not(any('i' in sublist for sublist in keyMatrix))):
        part[0] = 'j'
        
    elif((part[0] == 'j') and not(any('j' in sublist for sublist in keyMatrix))):
        part[0] = 'i'
        
    if((part[1] == 'i') and not(any('i' in sublist for sublist in keyMatrix))):
        part[1] = 'j'
        
    elif((part[1] == 'j') and not(any('j' in sublist for sublist in keyMatrix))):
        part[1] = 'i'
        
    x1 , y1 = np.where(keyMatrix == part[0])
    x2 , y2 = np.where(keyMatrix == part[1])
            
    if(x1 == x2):
        cpart[0] = keyMatrix[x1 , y1%4 if y1 == 4 else y1+1]
        cpart[1] = keyMatrix[x2 , y2%4 if y2 == 4 else y2+1]
                
    elif(y1 == y2):
        cpart[0] = keyMatrix[x1%4 if x1 == 4 else x1+1 , y1]
        cpart[1] = keyMatrix[x2%4 if x2 == 4 else x2+1 , y2]
                
    else:
        cpart[0] = keyMatrix[x1 , y2]
        cpart[1] = keyMatrix[x2 , y1]

    return cpart[0] , cpart[1]
```

```python
playfair_plaintext = readTextFile("Input Files\PlayFair\playfair_plain.txt")

plain_text_list = list(playfair_plaintext)
cipher_text = ""
kMatrix = []

keywords = ["rats" , "archangel"]
for i in range(len(keywords)):
    matrix = keyMatrix((keywords[i].lower()).replace(" " , "") , 5)
    kMatrix.append(matrix)

for q in range(len(kMatrix)):
    print("PlayFair Encryption with key = " + keywords[q] + "\n")
    for i in range(len(plain_text_list)):
        plain_text = list((plain_text_list[i].lower()).replace(" " , ""))

        for j in range(0 , len(plain_text) , 2):
            if(plain_text[j] == plain_text[j+1]):
                plain_text.insert(j+1,'x')

        if(len(plain_text)%2 == 1):
            plain_text.insert(len(plain_text),'x')

        for k in range(0,len(plain_text),2):
            p1,p2 = playFairCipher(plain_text[k:k+2] , kMatrix[q])
            p1 = ", ".join(p1)
            p2 = ", ".join(p2)
            cipher_text += p1 + p2

        with open('PlayFair Encryption\\' + keywords[q]+'.txt', 'a') as txt:
            txt.write(cipher_text+"\n")
            print("Plain text: " + plain_text_list[i] + " ==> Cipher text: " + cipher_text +'\n')
        cipher_text = ""
    print("----------------\n")
             
print("Encryption completed successfully.")
```
